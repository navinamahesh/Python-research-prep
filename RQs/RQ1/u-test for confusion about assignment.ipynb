```python
#extracting transfer status and DeID:
transferStatusWithDeID = pre_survey[["transfer_student_status","DeID"]]
transferStatusWithDeID = transferStatusWithDeID.dropna() #dropping any undefined values
transferStatusWithDeID
```

```python
#extracting the relevant PA2 and PA9 columns:
PA2_confusion = PA2_data[["PA2_confusion","DeID"]]
PA3_confusion = PA3_data[["PA3_confusion","DeID"]]
PA4_confusion = PA4_data[["PA4_confusion","DeID"]]
PA5_confusion = PA5_data[["PA5_confusion","DeID"]]
PA6_confusion = PA6_data[["PA6_confusion","DeID"]]
PA7_confusion = PA7_data[["PA7_confusion","DeID"]]
PA8_confusion = PA8_data[["PA8_confusion","DeID"]]
PA9_confusion = PA9_data[["PA9_confusion","DeID"]]
```

```python
#creating a table with transfer status, deID and ALL PA confusion about assignment values:
finalTable = pd.merge(transferStatusWithDeID,PA2_confusion)
finalTable
```

```python
finalTable = pd.merge(finalTable,PA3_confusion)
finalTable = pd.merge(finalTable,PA4_confusion)
finalTable = pd.merge(finalTable,PA5_confusion)
finalTable = pd.merge(finalTable,PA6_confusion)
finalTable = pd.merge(finalTable,PA7_confusion)
finalTable = pd.merge(finalTable,PA8_confusion)
finalTable = pd.merge(finalTable,PA9_confusion)
finalTable
```

```python
#filtering the finalTable further into a table with only transfers and only natives:
finalTransfersTable = finalTable[finalTable.transfer_student_status == "Yes"]
finalTransfersTable.info() #17 transfer students that filled out every weekly PA reflection

#replacing 1(Not at all) with 1 and replacing 5(significantly) with 5:

for PA in ["PA2_confusion","PA3_confusion","PA4_confusion","PA5_confusion","PA6_confusion","PA7_confusion","PA8_confusion","PA9_confusion"]:
    finalTransfersTable.loc[finalTransfersTable[PA] == "1 (Not at all)", PA] = 1
    finalTransfersTable.loc[finalTransfersTable[PA] == "5 (Significantly)", PA] = 5
    
finalTransfersTable #viewing the table
```

```python
finalNativesTable = finalTable[finalTable.transfer_student_status == "No"]
finalNativesTable.info() #200 native students that filled every PA weekly reflection

for PA in ["PA2_confusion","PA3_confusion","PA4_confusion","PA5_confusion","PA6_confusion","PA7_confusion","PA8_confusion","PA9_confusion"]:
    finalNativesTable.loc[finalNativesTable[PA] == "1 (Not at all)", PA] = 1
    finalNativesTable.loc[finalNativesTable[PA] == "5 (Significantly)", PA] = 5
    
finalNativesTable
```

```python
#filter PA2 values for transfers and PA2 values for natives and store this in 2 different lists
#repeat this process for the rest of the PAs
#PA2_transfers_list = finalTransfersTable["PA2_confusion"].tolist()
#PA2_natives_list = finalNativesTable["PA2_confusion"].tolist()
#PA2_transfers_list

for PA in ["PA2_confusion","PA3_confusion","PA4_confusion","PA5_confusion","PA6_confusion","PA7_confusion","PA8_confusion","PA9_confusion"]:
    transfers_confusion_list = finalTransfersTable[PA].tolist() #getting column values and putting them in a list
    transfers_confusion_list = [int(i) for i in transfers_confusion_list] #converting everything to int to be safe
    
    natives_confusion_list = finalNativesTable[PA].tolist()
    natives_confusion_list = [int(i) for i in natives_confusion_list] #same process as above but for natives
    
    u_statistic, pvalue = stats.mannwhitneyu(transfers_confusion_list,natives_confusion_list) #conducting the actual u-test
    print(pvalue)
```
